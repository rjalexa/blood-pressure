### This very simple notebook gets a blood pressure readings file (in the provided dataset as generated by an OMRON blood pressure contraption) and generates descriptive statistics and related box/swarmplots to illustrate how the blood pressure behaves in the morning, mid-day and afternoon (as set by the arbitrary AM_LIMIT and PM_LIMIT constants and added as a "partofday" column to the dataframe

```python
import pandas as pd
import seaborn as sns
import datetime as dt
```

```python
AM_LIMIT = 10
PM_LIMIT = 16
LOGNAME = 'data/bp_log.csv'
```

```python
def partofday(hour):
    if hour <= AM_LIMIT:
        return 'AM'
    elif hour > AM_LIMIT and hour <= PM_LIMIT:
        return 'MID'
    else:
        return 'PM'
```

# read the data

```python
df = pd.read_csv(LOGNAME)
```

```python
df.head()
```

## now add this new "partofday" column to the dataframe via my partofday function

```python
df['partofday'] = df.apply(lambda row: partofday(pd.to_datetime(row['Measurement Date']).hour), axis=1)
```

## Here are simple descriptive stats of the BP values as per section of the day

```python
df.loc[df['partofday'] == 'AM'].describe()
```

```python
df.loc[df['partofday'] == 'MID'].describe()
```

```python
df.loc[df['partofday'] == 'PM'].describe()
```

# Now plot boxplots and datapoints of both Systolic and Diastolic BP values

```python
sns.set_theme(style="whitegrid")
```

```python
ax = sns.boxplot(x=df["partofday"], y=df["SYS"],hue=df["partofday"],data=df, order=["AM","MID","PM"])
ax = sns.swarmplot(x=df["partofday"], y=df["SYS"],hue=df["partofday"],data=df, order=["AM","MID","PM"])
ax.legend_.remove()
```

```python
ax = sns.boxplot(x=df["partofday"], y=df["DIA"],hue=df["partofday"],data=df, order=["AM","MID","PM"])
ax = sns.swarmplot(x=df["partofday"], y=df["DIA"],hue=df["partofday"],data=df, order=["AM","MID","PM"])
ax.legend_.remove()
```
